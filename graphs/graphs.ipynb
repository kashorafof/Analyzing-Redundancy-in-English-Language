```python
#dictionary to histogram
import matplotlib.pyplot as plt
# maps each rule to a file name
File_names = {
    'Coordinating Conjunction' : 'Coordinating_conjunction',
    'Abbreviation' : 'Abbreviation',
    'Phrasal Verb': 'Phrasal_Verb',
    'Subordinating Conjunction' : 'Subordinating_conjunction',
    'Noun and Its Identifier' : 'Noun_and_its_identifier',
}
```

```python
#import plt
from matplotlib.pyplot import figure
from scipy.stats import sem
import numpy as np

# graoh a word and a dictionary of its occurrences
def graph(word, occurrences):
    x= plt
    # calculate the error and standard deviation
    error = sem(list(occurrences.values()))
    standard_deviation = '{:.2f}'.format(np.std(list(occurrences.values())))
    
    # show the bars and the error
    x.bar(range(len(occurrences)), list(occurrences.values()), yerr = error, tick_label=list(occurrences.keys()), alpha = 0.75,capsize = 3)
    # set the x axis margin
    x.Axes.set_xmargin(plt.gca(), 0.02)
    # rotate the x axis text with angle 45
    x.xticks(rotation=30)
    #change font size of x axis
    x.tick_params(axis='x', labelsize=13)
    
    # set title size and make it bold
    #x.title(word, fontsize=15, fontweight='bold' )
    x.title(word, fontsize=17, loc = 'left')
    #x.xlabel('categories')
    x.ylabel('Occurrences per 100 word', fontsize=14)
    # set fig text tot segma symbol and the standard deviation
    x.figtext(0.76, 0.9,r'$\sigma$ = '+str(standard_deviation), fontsize=13)

    return x
```

```python

dir = 'scrapped/'
df = pd.read_csv(dir + 'result.csv', index_col=0)
# dataframe to dictionary 
df_dict = df.to_dict()
for rule,occ in df_dict.items():
    x = graph(rule,occ)
    x.savefig(dir + 'eps/' + File_names[rule] + '.eps', bbox_inches='tight', dpi=300)
    #modify the dimension of savefig
    x.savefig(dir + 'png/'+ File_names[rule] + '.png', bbox_inches='tight', dpi=300)
    #clear plt
    #x.show()
    plt.clf()
```
